### Use the apply function to do a advance transformation of data

```python
def transform_country(country):
    if country == "USA":
        return "United States"
    else:
        return country
```

```python
transform_country("Brazil")
```

```python
transform_country("USA")
```

```python
df['Country'].apply(transform_country)
```
